# Association Rule Mining, Recommendation Systems and Dimensionality Reduction

```python
import pandas as pd
import os
os.getcwd()
```

```python
dataset = pd.read_excel('Online_Retail.xlsx')
```

```python
len(dataset)
```

```python
#Strip unnecessary sapcing from Description column
dataset['Description'] =dataset['Description'].str.strip()
```

```python
dataset.info()
```

```python
#delete few records as the data is huge
dataset['InvoiceNo']=dataset['InvoiceNo'].astype('str')
dataset=dataset[~dataset['InvoiceNo'].str.contains('C')]
```

```python
len(dataset)
```

```python
#Has to use only France data
```

```python
basket = (dataset[dataset['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets
```

```python
basket_sets.to_csv('basket.csv')
```

```python
pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itms = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itms, metric='lift', min_threshold=8)
```

```python
len(rules)
```

# K-Fold Validation

```python
#K-Fold Validation - train/test - train data is further broken and model is applied to each broken train data.
#create multiple samples, run models on each samples, get accuracy for each samples and acc gives us a range (20%-95% or 90% - 97%)
#Additional check on all my models (Classification and regression)
```

```python
import os
os.getcwd()
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
x = data.iloc[:,[2,3]].values
```

```python
y = data.iloc[:,4].values
```

```python
from sklearn.preprocessing import StandardScaler
sc_x =StandardScaler()
x=sc_x.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=10)
```

```python
from sklearn.svm import SVC
classifier_svm_sig = SVC(kernel = 'sigmoid')
classifier_svm_sig.fit(x_train,y_train)
```

```python
y_pred = classifier_svm_sig.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier_svm_sig, X = x_train, y = y_train, cv=10)
```

```python
accuracies
```

```python
#What is the average accuracy?
accuracies.mean()
```

```python
from sklearn.svm import SVC
classifier_svm_rbf = SVC(kernel = 'rbf')
classifier_svm_rbf.fit(x_train,y_train)
```

```python
y_pred = classifier_svm_rbf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier_svm_rbf, X = x_train, y = y_train, cv=10)
```

```python
accuracies
```

```python
accuracies.mean()
```

```python
y = data.iloc[:,6].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.decomposition import PCA
#Principal component analysis
pca = PCA(n_components=2)
x_train_pca_2 = pca.fit_transform(x_train)
x_test_pca_2 = pca.fit_transform(x_test)
```

```python
#Performance wise 
import time
size = 1000000
l1 = range(size)
l2 = range(size)
a1 = np.arange(size)
a2 = np.arange(size)

start = time.time()
result = [(x,y) for x , y in zip(l1,l2)]
print('Time taken by list')
print((time.time()-start)*1000)

start = time.time()
result = a1 + a2
print('Time taken by Array')
print((time.time()-start)*1000)

```

```python
#SVM - Linear
from sklearn.svm import SVC
classifier_svc_lin = SVC(kernel = 'linear')
classifier_svc_lin.fit(x_train, y_train)
```

```python
y_pred_svc_lin = classifier_svc_lin.predict(x_test)
```

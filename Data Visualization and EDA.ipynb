```python
year = [2015,2016,2017,2018,2019]
sales = [4543,4567,7656,5566,5544]
profit = [454,767,566,443,344]
```

```python
plt.plot(year,sales)
plt.plot(year,profit)
plt.title('Sales and Profit vs Year')
plt.xlabel('Years')
plt.ylabel('Sales and Profit')
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(year,sales, 'Red')
plt.title('Sales vs Year')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.subplot(3,1,2)
plt.plot(year,profit,'Blue')
plt.title('Profit vs Year')
plt.xlabel('Years')
plt.ylabel('Profit')
plt.subplot(3,1,3)
plt.plot(year,sales, 'Red')
plt.title('Sales vs Year')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.subplot(3,1,2)
plt.plot(year,profit,'Blue')
plt.title('Profit vs Year')
plt.xlabel('Years')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
plt.tight_layout()
plt.show()
```

```python
sales = [4543,4567,7656,5566,5544]
profit = [454,767,566,443,344]
plt.scatter(y=profit, x = sales)
plt.annotate('Customer5', xy=(5544,344))
plt.style.use('seaborn-whitegrid')
```
